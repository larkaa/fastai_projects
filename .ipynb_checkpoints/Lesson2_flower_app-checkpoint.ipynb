```python
#for flower data

tfms = get_transforms(max_rotate=50,flip_vert=True)
path = Path('/home/larkaa/.fastai/data/flower_data/')
np.random.seed(42)
data = ImageDataBunch.from_folder(path,train='train',valid='valid',test='test', #valid_pct=0.2,
        ds_tfms=tfms, size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.show_batch(rows=2, figsize=(7,8))
```

```python
data.c, len(data.train_ds), len(data.valid_ds),len(data.test_ds)
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)

s2 = 'flowers-stage-2'
#learn.save(s2);
learn.load(s2);
```

```python
learn.lr_find();
learn.recorder.plot()
```

```python
learn.fit_one_cycle(1, max_lr=slice(1e-5,1e-3))
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
interp.most_confused(min_val=2)
```

```python
img1 = open_image('/home/larkaa/Downloads/thorn_apple_test.jpg')
#img2 = open_image('/content/HADWFK.jpg')
img1.show()
pred_class,pred_idx,outputs = learn.predict(img1)
print(pred_class)

with open('/home/larkaa/.fastai/data/flower_data/cat_to_name.json') as json_file:  
    flower_names = json.load(json_file)
print(flower_names[str(pred_class)])
```
